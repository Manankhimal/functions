### 1. Types of Data
Data is categorized into **qualitative** (categorical) and **quantitative** (numerical).

- **Qualitative Data:** Non-numeric and descriptive.
  - **Nominal Scale:** Categories with no order (e.g., colors, gender).
  - **Ordinal Scale:** Categories with order but no fixed interval (e.g., satisfaction levels: good, average, bad).

- **Quantitative Data:** Numeric and measurable.
  - **Interval Scale:** Ordered, with equal intervals but no true zero (e.g., temperature in Celsius).
  - **Ratio Scale:** Ordered, equal intervals, and a true zero (e.g., height, weight, age).

---

### 2. Measures of Central Tendency
- **Mean (Average):** Used when data is continuous and normally distributed.
  - Example: Average income in a company.
- **Median (Middle Value):** Used when data has outliers.
  - Example: Median house prices.
- **Mode (Most Frequent Value):** Used for categorical data.
  - Example: Most popular car color.

---

### 3. Concept of Dispersion
- **Variance:** Measures how data points deviate from the mean.
- **Standard Deviation:** Square root of variance, providing a clearer measure of spread.

Example: In test scores, higher standard deviation means more variation in performance.

---

### 4. Box Plot and Distribution Insights
- A **box plot** visualizes data spread using quartiles.
- It shows **median, interquartile range (IQR), outliers, and skewness**.

---

### 5. Role of Random Sampling
- Ensures unbiased representation of a population.
- Improves **generalizability** of study results.
- Example: Conducting surveys using random selection of participants.

---

### 6. Skewness and Its Types
- **Positive Skew:** Right tail longer (e.g., income distribution).
- **Negative Skew:** Left tail longer (e.g., test scores).
- **Effect:** Skewness affects mean and median alignment.

---

### 7. Interquartile Range (IQR) and Outlier Detection
- **IQR = Q3 - Q1 (Middle 50% of data).**
- Outliers: **Below Q1 - 1.5(IQR) or Above Q3 + 1.5(IQR).**
- Example: Detecting unusual test scores.

---

### 8. Conditions for Binomial Distribution
- Fixed number of **n trials**.
- Two outcomes: **Success or failure**.
- **Constant probability** of success.
- Independent trials.

Example: Flipping a coin 10 times.

---

### 9. Properties of Normal Distribution and Empirical Rule
- Bell-shaped, symmetric around mean.
- **Empirical Rule:**
  - **68%** within 1 standard deviation.
  - **95%** within 2 standard deviations.
  - **99.7%** within 3 standard deviations.

Example: Human heights.

---

### 10. Poisson Process Example
- Used for rare, random events occurring in a fixed time/space.
- Example: Number of customers arriving at a store per hour.
- Probability: **P(X = k) = (λ^k * e^(-λ)) / k!**, where **λ** is the mean occurrence rate.

---

### 11. Random Variables
- **Discrete:** Countable outcomes (e.g., dice roll).
- **Continuous:** Infinite values within a range (e.g., height, weight).

---

### 12. Covariance and Correlation Calculation
- **Covariance:** Measures direction of relationship.
  - **Positive**: Increase together.
  - **Negative**: One increases, the other decreases.
- **Correlation (r):** Standardized covariance (-1 to 1).
  - **r = 1** (Strong positive).
  - **r = -1** (Strong negative).
  - **r = 0** (No correlation).
