```python
### 1. **What is a random variable in probability theory?**
A random variable is a numerical outcome of a random phenomenon. It assigns a real number to each outcome of a probabilistic experiment.

---

### 2. **What are the types of random variables?**
Random variables are classified into:
- **Discrete Random Variable**: Takes a countable number of distinct values.
- **Continuous Random Variable**: Takes an infinite number of values within a given range.

---

### 3. **What is the difference between discrete and continuous distributions?**
- **Discrete Distribution**: Deals with countable outcomes (e.g., number of heads in coin flips).
- **Continuous Distribution**: Deals with an infinite set of possible values (e.g., height of people in a population).

---

### 4. **What are probability distribution functions (PDF)?**
- For **discrete** random variables: The Probability Mass Function (PMF) gives the probability of each distinct outcome.
- For **continuous** random variables: The Probability Density Function (PDF) describes the relative likelihood of different values.

---

### 5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
- **PDF (or PMF)** gives the probability of a specific value.
- **CDF** gives the probability that the random variable is less than or equal to a certain value.

---

### 6. **What is a discrete uniform distribution?**
A discrete uniform distribution assigns equal probability to a finite number of outcomes (e.g., rolling a fair die: each face has a 1/6 probability).

---

### 7. **What are the key properties of a Bernoulli distribution?**
- It models a **single** experiment with two possible outcomes: success (\(1\)) or failure (\(0\)).
- The probability of success is \( p \), and failure is \( 1 - p \).
- Mean = \( p \), Variance = \( p(1-p) \).

---

### 8. **What is the binomial distribution, and how is it used in probability?**
- Models the number of successes in \( n \) independent Bernoulli trials.
- Probability mass function:
  \[
  P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
  \]
- Used in quality control, medical trials, and risk assessment.

---

### 9. **What is the Poisson distribution and where is it applied?**
- Models the number of events occurring in a fixed interval of time/space given a constant rate \( \lambda \).
- Formula:
  \[
  P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
  \]
- Used in queuing systems, traffic flow, and call centers.

---

### 10. **What is a continuous uniform distribution?**
A distribution where all outcomes in a range \([a, b]\) have equal probability.
PDF:
\[
f(x) = \frac{1}{b-a}, \quad a \leq x \leq b
\]

---

### 11. **What are the characteristics of a normal distribution?**
- Bell-shaped and symmetric about the mean.
- Defined by mean (\(\mu\)) and standard deviation (\(\sigma\)).
- 68-95-99.7 rule: 68% of data falls within 1\(\sigma\), 95% within 2\(\sigma\), and 99.7% within 3\(\sigma\).

---

### 12. **What is the standard normal distribution, and why is it important?**
- A normal distribution with mean \( \mu = 0 \) and standard deviation \( \sigma = 1 \).
- Used for **Z-scores** and hypothesis testing.

---

### 13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
- States that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases.
- Enables inference on population parameters even when the original distribution is not normal.

---

### 14. **How does the Central Limit Theorem relate to the normal distribution?**
- Regardless of the population distribution, sample means tend to follow a normal distribution if the sample size is large enough (\( n \geq 30 \)).

---

### 15. **What is the application of Z statistics in hypothesis testing?**
- Used to test population means when the sample size is large (\( n > 30 \)) and population variance is known.
- Helps determine if observed differences are statistically significant.

---

### 16. **How do you calculate a Z-score, and what does it represent?**
Formula:
\[
Z = \frac{X - \mu}{\sigma}
\]
It measures how many standard deviations a value \( X \) is from the mean.

---

### 17. **What are point estimates and interval estimates in statistics?**
- **Point Estimate**: A single value estimate of a population parameter (e.g., sample mean).
- **Interval Estimate**: A range within which the parameter is expected to lie (e.g., confidence interval).

---

### 18. **What is the significance of confidence intervals in statistical analysis?**
They provide a range of values that likely contain the population parameter with a given confidence level (e.g., 95%).

---

### 19. **What is the relationship between a Z-score and a confidence interval?**
- A Z-score determines the margin of error in confidence intervals.
- Confidence Interval:
  \[
  \bar{X} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
  \]

---

### 20. **How are Z-scores used to compare different distributions?**
- Standardizes different distributions, allowing direct comparison of values from different datasets.

---

### 21. **What are the assumptions for applying the Central Limit Theorem?**
- Sample size should be large (\( n \geq 30 \)).
- Samples should be independent.
- Data should not be highly skewed.

---

### 22. **What is the concept of expected value in a probability distribution?**
The **expected value (mean)** of a random variable represents its long-run average:
\[
E[X] = \sum x P(X = x) \quad \text{(Discrete)}
\]
\[
E[X] = \int x f(x) dx \quad \text{(Continuous)}
\]

---

### 23. **How does a probability distribution relate to the expected outcome of a random variable?**
- The probability distribution defines the likelihood of each possible outcome, and the expected value gives the average outcome over many trials.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2. Discrete Uniform Distribution
n = 10
discrete_uniform = np.random.randint(1, n+1, size=1000)
plt.hist(discrete_uniform, bins=n, density=True, alpha=0.6, color='g')
plt.title("Discrete Uniform Distribution")
plt.show()
```

```python
#3
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)
print("Bernoulli PDF (p=0.5, x=1):", bernoulli_pdf(0.5, 1))
```

```python
# 4. Binomial Distribution Simulation
n, p = 10, 0.5
binomial_samples = np.random.binomial(n, p, 1000)
plt.hist(binomial_samples, bins=10, density=True, alpha=0.6, color='b')
plt.title("Binomial Distribution")
plt.show()
```

```python
# 5. Poisson Distribution
poisson_samples = np.random.poisson(lam=5, size=1000)
plt.hist(poisson_samples, bins=10, density=True, alpha=0.6, color='r')
plt.title("Poisson Distribution")
plt.show()
```

```python
# 6. Continuous Uniform Distribution
continuous_uniform = np.random.uniform(0, 1, 1000)
plt.hist(continuous_uniform, bins=20, density=True, alpha=0.6, color='c')
plt.title("Continuous Uniform Distribution")
plt.show()
```

```python
# 7. Normal Distribution & Histogram
normal_samples = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(normal_samples, bins=20, density=True, alpha=0.6, color='m')
plt.title("Normal Distribution")
plt.show()
```

```python
# 8. Z-Score Calculation
def calculate_z_score(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return [(x - mean) / std_dev for x in data]
z_scores = calculate_z_score(normal_samples)
plt.hist(z_scores, bins=20, density=True, alpha=0.6, color='y')
plt.title("Z-Score Distribution")
plt.show()
```

```python
# 9. Central Limit Theorem
sample_means = [np.mean(np.random.normal(0, 1, 30)) for _ in range(1000)]
plt.hist(sample_means, bins=20, density=True, alpha=0.6, color='k')
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

```python
# 10. Confidence Interval Calculation
data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std_err = stats.sem(data)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)
print(f"95% Confidence Interval: {conf_interval}")
```

```python
#11
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
num_samples = 1000  # Number of samples
sample_size = 50     # Size of each sample

# Generate samples from a uniform distribution (not normal)
population = np.random.uniform(0, 10, 10000)

# Compute sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plot
sns.histplot(sample_means, kde=True, bins=30)
plt.title("Sampling Distribution of the Mean (CLT Verification)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#12
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label="Standard Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Mean=0, Std=1)")
plt.legend()
plt.show()
```

```python
#13
from scipy.stats import binom

n, p = 10, 0.5  # 10 trials, 50% probability of success
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.bar(x, y)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
#14
def z_score(x, mean, std_dev):
    return (x - mean) / std_dev

data = np.random.normal(50, 10, 100)
z = z_score(60, np.mean(data), np.std(data))

print(f"Z-score: {z}")
```

```python
#15
from scipy.stats import norm

# Sample Data
sample_mean = 52
pop_mean = 50
std_dev = 5
n = 30  # Sample size

# Z-score calculation
z_stat = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z_stat)

print(f"Z-score: {z_stat}, P-value: {p_value}")
```

```python
#16
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
confidence = 0.95
mean = np.mean(data)
std_err = stats.sem(data)

margin = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * std_err
conf_interval = (mean - margin, mean + margin)

print(f"95% Confidence Interval: {conf_interval}")
```

```python
#17
sample = np.random.normal(10, 3, 100)
mean = np.mean(sample)
std_dev = np.std(sample, ddof=1)
n = len(sample)

margin_of_error = 1.96 * (std_dev / np.sqrt(n))  # 95% CI
conf_interval = (mean - margin_of_error, mean + margin_of_error)

print(f"Confidence Interval: {conf_interval}")
```
