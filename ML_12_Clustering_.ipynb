```python
#31 Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with  decision boundaries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Step 1: Generate synthetic overlapping data
X, y_true = make_blobs(n_samples=400, centers=3, cluster_std=2.5, random_state=42)  # high std for overlap

# Step 2: Fit K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Step 3: Define function to plot decision boundaries
def plot_decision_boundaries(X, model, ax):
    h = 0.1  # mesh step size
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    ax.contourf(xx, yy, Z, alpha=0.3, cmap=cmap_light)
    scatter = ax.scatter(X[:, 0], X[:, 1], c=labels, cmap=cmap_bold, edgecolor='k', s=50)
    ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
               c='yellow', s=200, marker='X', label='Centroids', edgecolor='k')
    ax.set_title('K-Means Clustering with Decision Boundaries')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.legend()

# Step 4: Plot
fig, ax = plt.subplots(figsize=(8, 6))
plot_decision_boundaries(X, kmeans, ax)
plt.show()
```

```python
#43 Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN  side-by-side
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Step 1: Generate noisy circular data
X, y_true = make_circles(n_samples=300, factor=0.5, noise=0.07, random_state=42)

# Step 2: Scale the data for DBSCAN (helps with distance-based clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Step 4: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Step 5: Plot results side-by-side
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# KMeans plot
axs[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', edgecolor='k', s=50)
axs[0].set_title('KMeans Clustering')
axs[0].set_xlabel('Feature 1')
axs[0].set_ylabel('Feature 2')
axs[0].grid(True)

# DBSCAN plot
# Mark noise points in red (label = -1)
colors = ['red' if label == -1 else f'C{label}' for label in dbscan_labels]
axs[1].scatter(X[:, 0], X[:, 1], c=colors, edgecolor='k', s=50)
axs[1].set_title('DBSCAN Clustering')
axs[1].set_xlabel('Feature 1')
axs[1].set_ylabel('Feature 2')
axs[1].grid(True)

plt.suptitle('Clustering on Noisy Circular Data: KMeans vs DBSCAN', fontsize=16)
plt.show()
```

```python
#44  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler

# Load Iris dataset
iris = load_iris()
X = iris.data

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Compute Silhouette Coefficient for each sample
silhouette_vals = silhouette_samples(X_scaled, cluster_labels)

# Sort samples by cluster for better visualization
y_lower = 10
fig, ax = plt.subplots(figsize=(8, 6))

for i in range(3):
    # Aggregate silhouette scores for samples in cluster i
    ith_cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.nipy_spectral(float(i) / 3)
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, ith_cluster_silhouette_vals,
                     facecolor=color, edgecolor=color, alpha=0.7)

    # Label the silhouette plots with cluster numbers in the middle
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    y_lower = y_upper + 10  # 10 for spacing between clusters

ax.set_title("Silhouette Plot for KMeans Clustering on Iris Dataset")
ax.set_xlabel("Silhouette Coefficient Values")
ax.set_ylabel("Cluster Label")

# Draw vertical line for average silhouette score
avg_score = silhouette_score(X_scaled, cluster_labels)
ax.axvline(x=avg_score, color="red", linestyle="--", label=f'Avg Score = {avg_score:.2f}')

ax.set_yticks([])  # Clear y-axis labels/ticks
ax.set_xlim([-0.1, 1])
ax.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
#46  Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4  features)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Load Wine dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

# Create DataFrame with first 4 features
df = pd.DataFrame(X[:, :4], columns=feature_names[:4])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df)

# Plot pairplot with cluster assignments as hue
sns.pairplot(df, hue='cluster', palette='Set1', diag_kind='kde')
plt.suptitle('KMeans Cluster Assignments on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

```python
#47  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the  count
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate noisy blobs
X, y_true = make_blobs(n_samples=400, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

# Scale features for DBSCAN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (exclude noise label -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
noise_count = list(labels).count(-1)

print(f"Number of clusters found by DBSCAN: {n_clusters}")
print(f"Number of noise points identified by DBSCAN: {noise_count}")
```

```python
#48  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Load Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions using t-SNE
tsne = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto')
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=10)
labels = agglo.fit_predict(X_tsne)

# Plot clusters
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50, edgecolor='k')
plt.title('Agglomerative Clustering on Digits Dataset (t-SNE reduced)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(scatter, ticks=range(10), label='Cluster Label')
plt.grid(True)
plt.show()
```
